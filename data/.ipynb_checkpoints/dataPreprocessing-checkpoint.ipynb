```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
feature = pd.read_csv('feature.csv', index_col=0)
```

## Data Visualization

```python
missingColumns = np.sum(feature.isnull(),axis=0)
```

```python
feature.index = pd.to_datetime(feature.index)
```

```python
f, ax = plt.subplots(figsize = (14, 5))
plt.hist(missingColumns, 100)
plt.xlabel('Missing Count')
plt.ylabel('Sample Count')
plt.show()
```

## Fix the Missing Value

+ Cut the NAN value in the head and tail of time series
+ Sample the data with daily frequency
+ Fix the missing value with the linear method

```python
AllSampled = []
for index in range(feature.shape[1]):
    thisData = feature.iloc[:,index]
    # thisData = thisData.replace('', np.NaN)
    first_idx = thisData.first_valid_index()
    last_idx = thisData.last_valid_index()
    # Remove the begining and the end invalid value of the time series
    thisData = thisData.loc[first_idx:last_idx]
    # Up sample the annual data into daily with quadratic method
    upsampled = thisData.resample('D', convention='start').asfreq()
    upsampled = upsampled.astype(float).interpolate(method='linear')
    # Append the sampled data
    AllSampled.append(pd.DataFrame(upsampled))
```

```python
AllSampledConcat = pd.concat(AllSampled, axis=1)
```

```python
AllSampledConcat.index
```

```python
# Delete the data which is after the date of 2018-9-29
stripedData = AllSampledConcat.loc[AllSampledConcat.index < pd.to_datetime('2018-09-29'),:]
```

```python
f, ax = plt.subplots(figsize = (14, 5))
sns.set_style("whitegrid")
nanFeatures = np.sum(stripedData.isnull(), axis=0)
#ax = sns.barplot(nanFeatures.index, nanFeatures.values)
plt.hist(nanFeatures, 100)
plt.xlabel('Missing Count')
plt.ylabel('Sample Count')
plt.show()
```

## Filter the Feature with Missing Rate Higher than the Threshold (80%)

```python
stripedData = stripedData.loc[:, nanFeatures/stripedData.shape[0]<0.8]
```

```python
# Show the percentage of missing feature as time goes by. 
f, ax = plt.subplots(figsize = (14, 5))
sns.set_style("whitegrid")
plt.plot(np.sum(stripedData.isnull()/float(stripedData.shape[1])*100, axis=1))
plt.xlabel('Date')
plt.ylabel('Missing Feature Percentage')
plt.title('The Percentage of Missing Feature Curve')
plt.show()
```

In the plot, we can see that the data mainly locate in the 1981 to 2016

```python
stripedData = AllSampledConcat.loc[(AllSampledConcat.index > pd.to_datetime('1981-01-01')) & (AllSampledConcat.index < pd.to_datetime('2018-09-29')),:]
```

```python
# Show the percentage of missing feature as time goes by. 
f, ax = plt.subplots(figsize = (14, 5))
sns.set_style("whitegrid")
plt.plot(np.sum(stripedData.isnull()/float(stripedData.shape[1])*100, axis=1))
plt.xlabel('Date')
plt.ylabel('Missing Feature Percentage')
plt.title('The Percentage of Missing Feature Curve')
plt.show()
```

Replace the NAN values in the head and tail with the closest data.

```python
from tqdm import tqdm
for i in tqdm(range(stripedData.shape[1])):
    stripedData.ix[:stripedData.iloc[:,i].first_valid_index(), i] = \
                            stripedData.ix[stripedData.iloc[:,i].first_valid_index(),i]
    stripedData.ix[stripedData.iloc[:,i].last_valid_index():, i] = \
                            stripedData.ix[stripedData.iloc[:,i].last_valid_index(),i]
```
